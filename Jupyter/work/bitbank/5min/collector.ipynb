```python
import pickle
import urllib.request
import json
import pprint
import pandas as pd
import datetime
```

```python
type = '5min'
save_dir = './data/'
start = '20180601'
end = '20180920'
format = '%Y%m%d'
day = datetime.datetime.strptime(start, format)
end_day = datetime.datetime.strptime(end, format)
iteration = 0
```

```python
while iteration < 1000:
    day_str = day.strftime(format)
    
    # http
    url = 'http://192.168.99.100:10080/public/candlestick/btc_jpy/'+type+'?time='+day_str
    req = urllib.request.Request(url)
    with urllib.request.urlopen(req) as res:
        body = res.read()
    if body is None:
        print('fail to get json in '+day_str+'...')
        break
        
    # decode
    my_str = body.decode("utf-8")
    my_dict = json.loads(my_str)
    candlestick = my_dict['candlestick'][0]['ohlcv']
    # header
    df = pd.DataFrame(candlestick, columns=['open', 'high', 'low', 'end', 'turnover', 'time'])
    # timeformat
    df['time'] = pd.to_datetime(df['time'], unit='ms')
    
    # save
    save_file = save_dir+day_str+'_'+type+'.pkl'
    with open(save_file, 'wb') as f:
        print('saving'+day_str)
        pickle.dump(df, f)
    
    day += datetime.timedelta(days=1)
    if day >= end_day:
        break
    iteration += 1
    
print('end')
```
